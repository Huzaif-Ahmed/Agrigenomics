```python
import pandas as pd
import numpy as np
```

```python
import json
with open("clusters_dict.json", 'r') as f:
    data = json.load(f)
data
```

```python
df=pd.read_csv("merged_final24.csv")
df
```

```python
d=dict()
for c,i in data.items():
    dic=dict()
    df1 = df[df["Cultivar ID"].isin(i)]
    subpopulations=list(set(df1["Subpopulation"]))
    for x in subpopulations:
        df2 = df1[df1["Subpopulation"].isin([x])]
        minheight = min(df2["Plant Height (cm)"])
        maxheight = max(df2["Plant Height (cm)"])
        heightrange=str(maxheight)+"-"+str(minheight)
        dic[x]=heightrange
    minheight=min(df1["Plant Height (cm)"])
    maxheight = max(df1["Plant Height (cm)"])
    
    
    for j in i:
        d1 = dict()
        d1["sequence"] = df[df["Cultivar ID"]==j]["Sequence"].iloc[0]
        d1["Subpopulation"]=dic
        d1["minheight"]=minheight
        d1["maxheight"]=maxheight
        d[j]=d1

d
```

```python
with open('final_classification.json', 'w') as f:
    json.dump(d, f)
```
