```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("merged_final24.csv")
df

```

```python
chromosome_columns = ["ch" + str(i) for i in range(1, 25)]
new_df = pd.DataFrame(columns=chromosome_columns)


for index, row in df.iterrows():
    sequence = row["Sequence"]
    
    
    chromosomes = list(sequence)
    
    
    new_df.loc[index] = chromosomes


df_with_chromosomes = pd.concat([df, new_df], axis=1)
```

```python
df=df_with_chromosomes
df
```

```python
del df["Sequence"],df["Cultivar ID"]
df
```

```python
label1=df["Plant Height (cm)"]
label2=df["Subpopulation"]
```

```python
del df["Plant Height (cm)"],df["Subpopulation"]
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(df, label1, test_size=0.8, shuffle=True)
```

# ANN for height

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python


model = Sequential()
model.add(Dense(64, input_dim=24, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(x_train, y_train, epochs=100, batch_size=32)

# Make predictions
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score, recall_score, precision_score, confusion_matrix


print("Accuracy: ", accuracy_score(y_test, y_pred))
```
